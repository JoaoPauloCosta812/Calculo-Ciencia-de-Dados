1. Marque quais desses métodos/algoritmos muito populares em ciência de dados são baseados no uso de derivada:

    1. Método Mínimos Quadrados
    2. Gradiente descendente
    3. Newton Raphson
    4. CART (Árvore de decisão)

**Método dos Mínimos Quadrados, Gradiente Descendente, Newton–Raphson.**

- **Mínimos Quadrados (OLS):** o ajuste é obtido *minimizando a soma dos quadrados dos resíduos*. A solução clássica (equações normais) vem de *anular o gradiente* da função de perda quadrática em relação aos parâmetros — isto é, usa derivadas.
- **Gradiente Descendente:** é, por definição, um *método de otimização via derivadas*, que segue a direção oposta ao gradiente para reduzir a função objetivo.
- **Newton–Raphson:** usa *derivadas de 1ª e 2ª ordem* (gradiente e Hessiana) para construir aproximações quadráticas e *encontrar raízes/ótimos* rapidamente.
- **CART (árvores de decisão):** não é baseado em derivadas; faz *particionamentos* do espaço de atributos via critérios como redução de impureza (Gini/entropia/MSE), sem cálculo de gradiente.


2. Dada uma base de dados com uma variável resposta $y$ e um conjunto de variáveis explicativas. Considere uma estrutura de um modelo de regressão. Explique com suas palavras por que não é possível obter parâmetros que forneçam um erro quadrático médio (EQM) menor que o obtido com estimadores de mínimos quadrados.

No mesmo modelo de regressão linear e usando a mesma função de perda quadrática, o estimador de Mínimos Quadrados (OLS) é, por construção, aquele que minimiza a Soma dos Quadrados dos Resíduos (SSE) sobre os parâmetros.  
Como o EQM de treinamento é proporcional à SSE (EQM = SSE / n, desconsiderando constantes), qualquer outro conjunto de parâmetros produzirá SSE (e portanto EQM) maior ou igual ao do OLS.

Intuição rápida: a função objetivo é convexa em relação aos coeficientes; o OLS encontra o mínimo global (onde o gradiente é zero). Sair desse ponto só pode aumentar a perda quadrática.

> Observação importante: sob as hipóteses do teorema de *Gauss–Markov*, o OLS é o melhor estimador linear não-viesado (BLUE), tendo menor variância entre os estimadores lineares não-viesados.  
> Entretanto, se considerarmos outros objetivos (ex.: regularização Ridge/Lasso) ou aceitarmos vieses (buscando menor erro de generalização), pode-se reduzir o erro esperado fora da amostra — mas não o EQM de treinamento para a mesma função quadrática.
